```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("fish.csv")
df
```

```python
df.info()
df.describe()
```

```python
df.groupby("Species").sum()

df.groupby("Species").mean()
```

```python
df.groupby("Species").agg([sum,np.mean])
```

```python
group_df = df.groupby("Species")
df.sort_values("Weight",ascending=False)
```

```python
df[["Species","Weight"]]
```

```python
df[df["Length1"]>40].sort_values("Length1")
```

```python
df.drop_duplicates(subset="Species")
```

```python
df.pivot_table(values="Weight",index="Species",aggfunc=np.mean).sort_values("Weight")
```

```python
df.groupby("Species")["Weight"].mean()
```

```python
weight_spec = df.groupby("Species").sum()["Weight"]
weight_spec = weight_spec.sort_values()
weight_spec
```

```python
weight_spec.plot(kind="bar",rot=60)
plt.show()
```

```python
df.plot(x="Species",y="Weight",kind="scatter")
df["Species"]
```

```python
bream = df[df["Species"]=="Bream"]
smelt = df[df["Species"]=="Smelt"]
parkki = df[df["Species"] == "Parkki"]
```

```python
bream["Length1"].hist(alpha=0.7)
smelt["Length1"].hist(alpha=0.7)
parkki["Length1"].hist(alpha=0.7)
plt.title("Length1 Histogram")
plt.legend(["Bream","Smelt","Parkki"])
plt.show()
```

```python
data = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(data,annot=True)
plt.show()
```
